```python
import numpy as np
import matplotlib.pyplot as plt

# cargar datos
data = np.loadtxt("CRITICAL1000.dat")

E = data[:, 0]
M = data[:, 1]

# tiempo real en MC steps
time = np.arange(len(E))

# -------------------------
# plots
# -------------------------

# E vs time (primeros 1000)
plt.figure()
plt.plot(time, E/10000)
plt.xlabel("Monte Carlo Step")
plt.ylabel("E/N")
plt.title("Energy evolution (first 1000 MCS)")
plt.tight_layout()
plt.savefig("Energy_critical_1000.png", dpi=600)

# M vs time (primeros 1000)
plt.figure()
plt.plot(time, M/10000)
plt.xlabel("Monte Carlo Step")
plt.ylabel("M/N")
plt.title("Magnetization evolution (first 1000 MCS)")
plt.tight_layout()
plt.savefig("Magnetization_critical_1000.png", dpi=600)


plt.show()
```
